#1. What is Power BI and why is it used in businesses?

Power BI is a Business Intelligence (BI) and Data Visualization tool developed by Microsoft. It helps users connect to different data sources, analyze data, and create interactive reports and dashboards.

Why it is used in businesses:

- To analyze large amounts of data.
- To create interactive dashboards and reports.
- To support better business decisions.
- To monitor Key Performance Indicators (KPIs).
- To share reports with teams securely.

#2. Name and explain the three main components of Power BI.

1. Power BI Desktop
- A Windows application used to import, clean, analyze, and visualize data.
- Used for creating reports.
2. Power BI Service
- A cloud-based online service.
- Used for publishing, sharing, and collaborating on reports and dashboards.
3. Power BI Mobile
- Mobile application for Android and iOS.
- Allows users to view reports and dashboards anywhere.

#3. Explain the Power BI workflow.

The Power BI workflow includes the following steps:

1. Connect to data sources.
2. Import or load data.
3. Clean and transform data using Power Query.
4. Create relationships between tables (if multiple tables exist).
5. Build charts, graphs, and dashboards.
6. Publish reports to Power BI Service.
7. Share reports with users.
8. Refresh data regularly.

#4.  List any four data cleaning tasks that can be performed in Power Query.

Any four data cleaning tasks are:

1. Remove duplicate records.
2. Remove blank or null values.
3. Change data types (Text, Number, Date, etc.).
4. Split or merge columns.

Other tasks include filtering rows, replacing values, renaming columns, and trimming spaces.

#6. Define Data View, Report View, and Model View. Explain the purpose of each view.

**Data View**
- Displays data in table format.
- Used to inspect and verify the loaded data.

**Report View**
- Used to create reports, charts, graphs, and dashboards.
- This is the main working area for report design.

**Model View**
- Displays relationships between tables.
- Used to create and manage table relationships.

#8. Split Owner Name to create two new columns as First Name and Last Name.

Steps:
1. Open Power Query Editor by clicking Transform Data.
2. Select the Owner Name column.
3. Go to Transform → Split Column → By Delimiter.
4. Choose Space as the delimiter.
5. Select Split at Left-most Delimiter.
6. Click OK.
7. Rename the new columns:
- First Name
- Last Name
8. Click Close & Apply to save the changes.

Example:
| Owner Name   | First Name | Last Name |
| ------------ | ---------- | --------- |
| Rahul Mehta  | Rahul      | Mehta     |
| Anita Sharma | Anita      | Sharma    |
| Neha Verma   | Neha       | Verma     |
